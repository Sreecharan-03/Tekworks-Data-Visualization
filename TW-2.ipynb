```python
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("Practice-Datasets .csv")
df
```

```python
# Create a histogram to analyze the distribution of Age
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Age',kde=True,bins=30,color='red',alpha=0.7,edgecolor='black')
```

```python
# 2. Plot a pie chart showing the percentage distribution of Gender.
plt.figure(figsize=(10,6))
plt.pie(df['Gender'].value_counts(),labels=['Male','Female'],colors=['blue','pink'],autopct='%1.1f%%',startangle=90,shadow=True)
```

```python
# 3. Create a bar chart showing the number of customers in each City.
plt.figure(figsize=(5,6))
sns.barplot(data=df,x='City',y='Customer_ID',color='orange',alpha=0.5)
plt.title('customers in each City')
```

```python
# 4. Plot a box plot to analyze Monthly Income distribution
plt.figure(figsize=(5,6))
sns.boxplot(data=df,x='Monthly_Income',color='black',width=0.6)
plt.title("Monthly Income distribution")
```

```python
# 5. Create a histogram of Spending Score and interpret the skewness.
plt.figure(figsize=(10,5))
sns.histplot(df['Spending_Score'],kde=True,bins=30,color='blue',edgecolor='black')
plt.title("Spending Score")
```

```python
# 6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.
plt.figure(figsize=(100,20))
sns.lineplot(data=df,x='Purchase_Date',y='Purchase_Amount',color='green',alpha=0.5,marker='*',markersize=7)
plt.title(' Amount trend over Purchase Date')
```

```python
# 7. Create a bar plot to compare average Monthly Income by Gender.
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Gender',y='Monthly_Income',color='purple',alpha=1)
plt.title('Monthly Income by Gender')
```

```python
# 9. Create a pie chart to show Product Category distribution.
plt.pie(df['Product_Category'].value_counts(),labels=['clothes','electronics','furniture','grocery','sports'],colors=['red','green','yellow','blue','orange'],autopct='%1.1f%%',startangle=90,shadow=True)
plt.title(' Product Category distribution')
```

```python
# 10. Plot a histogram of Purchase Amount with appropriate bin size.
sns.histplot(data=df,x='Purchase_Amount',bins=10,kde=True,color='red',alpha=0.5)
```

```python
# 11. Create a bar chart showing average Spending Score by City.
a=df.groupby('City')['Spending_Score'].mean().reset_index()
```

```python
a
```

```python
sns.barplot(data=a,x='City',y='Spending_Score',color='silver',alpha=0.7)
plt.title('average Spending Score by City')
```

```python
# 12. Plot a line chart showing total Purchase Amount per month.
df["Purchase_Date"]=pd.to_datetime(df["Purchase_Date"],errors='coerce',dayfirst=True)
df['month']=df['Purchase_Date'].dt.month
b=df.groupby('month')['Purchase_Amount'].sum().reset_index()
```

```python
b
```

```python
sns.lineplot(data=b,x='month',y='Purchase_Amount',color='gold',marker='^',markersize=8)
plt.title('total Purchase Amount per month')
```
